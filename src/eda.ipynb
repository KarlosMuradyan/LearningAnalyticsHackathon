```python
import pandas as pd
import ast
```

```python
course_data = pd.read_csv("data/ubc_course_calendar_data.csv")
```

```python
course_data['FACULTY'] = course_data['FACULTY'].str.replace('FACULTY OF ', '').str.strip()
course_data['COURSE_NUMBER'] = course_data['COURSE_NUMBER'].str.replace(r'[A-Z]', '').str.strip()
course_data['COURSE_CODE'] = course_data[['SUBJECT_CODE', 'COURSE_NUMBER']].apply(lambda x: ' '.join(x), axis=1)
course_data = course_data.dropna(subset=['FACULTY', 'COURSE_DESCRIPTION'])
```

```python
# filter course description and lecture type
course_data = course_data[course_data["SESSION_YEAR"] == "2018W"]
course_data = course_data[course_data["COURSE_DESCRIPTION"] != " "]
course_data = course_data[course_data["COURSE_DESCRIPTION"] != "Please see the Department website for further information on topics offered in the current session."]
course_data = course_data[course_data["COURSE_DESCRIPTION"] != "This is only a test."]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('seminar or lecture.', case = False)]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('May be taken more than once for credit.')]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('pass/fail', case = False)]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('Requires approval of a Department Head or Program Director.', case = False)]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('enrolment restrictions', case = False)]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('may be taken for credit', case = False)]
course_data = course_data[~course_data["COURSE_DESCRIPTION"].str.contains('permission of the department head', case = False)]
course_data = course_data[course_data["SECTION_TYPE"] == "LEC"]
```

```python
pd.set_option('max_colwidth', 4000)
course_data = course_data[["COURSE_CODE",
                           "COURSE_TITLE",
                           "COURSE_DESCRIPTION", 
                           "FACULTY",  
                           "CREDITS", 
                           "PRE_REQUISITE_DESCRIPTIONS"]].sort_values(by = ["COURSE_CODE"]).drop_duplicates()
```

```python
courses = {}

for index, row in course_data.iterrows():
    course = {'course_code': row['COURSE_CODE'],
              'course_title': row['COURSE_TITLE'],
              'faculty': row['FACULTY'],
              'credits': row['CREDITS'],
              'pre_requisites': row['PRE_REQUISITE_DESCRIPTIONS']}
    if courses.get(row['COURSE_DESCRIPTION']) is None:
        courses[row['COURSE_DESCRIPTION']] = [course]
    else:
        courses.get(row['COURSE_DESCRIPTION']).append(course)
```
